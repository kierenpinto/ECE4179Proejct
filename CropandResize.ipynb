```
images = os.listdir('/content/Train Imgs')
maps = os.listdir('/content/savedfiles')

images.sort()
maps.sort()

if len(images) != len(maps):
  print(len(images), len(maps))
```

```
image = cv2.imread('/content/Train Imgs/'+images[10])
MAP = cv2.imread('/content/savedfiles/'+images[10].split('.')[0]+'_classimg_nonconvex.png')

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(1,2,1)
ax.imshow(image)
ax = fig.add_subplot(1,2,2)
ax.imshow(MAP[:,:,0], cmap='Blues')

h, w, c = image.shape

horizontal_crops = int(w/512)
vertical_crops = int(h/512)

fig = plt.figure(figsize=(20,20))

counter = 0
for x in range(horizontal_crops):
  for y in range(vertical_crops):
    img_crop = copy.deepcopy(image[y*512:y*512+512, x*512: x*512+512, :])
    map_crop = copy.deepcopy(MAP[y*512:y*512+512, x*512: x*512+512,0])
    
    ax = fig.add_subplot(horizontal_crops*2, vertical_crops, counter*2+1)
    ax.imshow(img_crop)
    
    ax = fig.add_subplot(horizontal_crops*2, vertical_crops, counter*2+2)
    ax.imshow(map_crop, cmap='Blues')
    
    counter+=1
```

```
widths = []
heights = []

for image in images: 
  
  img = cv2.imread('/content/Train Imgs/'+image)
  
  h, w, c = img.shape
    
  if w not in widths:
    widths.append(w)
  
  if h not in heights:
    heights.append(h)
    
print(widths, heights)
```

```
import torchvision.transforms as Transforms
# Transforms.functional.RandomCrop(image, size = 5120, pad_if_needed=True)
# Transforms.functional.Resize(image, size = 512)

img_count=0
for image in images:
    img_count+=1
    print('Processing image {}/{}, number of images: {}'.format(img_count, len(images), len(os.listdir('/content/content/My Drive/ECE4179 - Machine Learning and Neural Networks/ECE4179 Project/Data/Image_crops'))))
    img = cv2.imread('/content/Train Imgs/'+image, -1)
    Maps = cv2.imread('/content/savedfiles/'+image.split('.')[0]+'_classimg_nonconvex.png', -1)
    img_crop = Transforms.functional.to_pil_image(img)
    img_crop = Transforms.CenterCrop(4608)(img_crop)
    img_crop = Transforms.functional.resize(img_crop, size = 512)
    map_crop = Transforms.functional.to_pil_image(Maps)
    map_crop = Transforms.CenterCrop(4608)(map_crop)
    map_crop = Transforms.functional.resize(map_crop, size = 512)
    
    img_name = '/content/content/My Drive/ECE4179 - Machine Learning and Neural Networks/ECE4179 Project/Data/Image_crops3/{}.png'.format(image.split('.')[0])
    map_name = '/content/content/My Drive/ECE4179 - Machine Learning and Neural Networks/ECE4179 Project/Data/Map_crops3/{}.png'.format(image.split('.')[0])

#     cv2.imwrite(img_name, img_crop)
    img_crop.save(img_name)
#     cv2.imwrite(map_name, map_crop) 
    map_crop.save(map_name)
```

```
number_of_classes = []
for image in images:
  img_name = '/content/content/My Drive/ECE4179 - Machine Learning and Neural Networks/ECE4179 Project/Data/Image_crops3/{}.png'.format(image.split('.')[0])
  map_name = '/content/content/My Drive/ECE4179 - Machine Learning and Neural Networks/ECE4179 Project/Data/Map_crops3/{}.png'.format(image.split('.')[0])
  img = cv2.imread(img_name, -1)
  Maps = cv2.imread(map_name, -1)
  # plt.figure()
  # plt.imshow(img)
  # plt.figure()
  # plt.imshow(Maps)
  number_of_classes.append((np.sum(Maps == 0),np.sum(Maps == 1),np.sum(Maps == 2),np.sum(Maps == 3)))
```
