```
import torchvision.transforms as Transforms
# Transforms.functional.RandomCrop(image, size = 5120, pad_if_needed=True)
# Transforms.functional.Resize(image, size = 512)

img_count=0
for image in images:
    img_count+=1
    print('Processing image {}/{}, number of images: {}'.format(img_count, len(images), len(os.listdir('/content/content/My Drive/ECE4179 - Machine Learning and Neural Networks/ECE4179 Project/Data/Image_crops'))))
    img = cv2.imread('/content/Train Imgs/'+image, -1)
    Maps = cv2.imread('/content/savedfiles/'+image.split('.')[0]+'_classimg_nonconvex.png', -1)
    img_crop = Transforms.functional.to_pil_image(img)
    img_crop = Transforms.functional.pad(img_crop,512,padding_mode='reflect')
    img_crop = Transforms.CenterCrop(5632)(img_crop)
    img_crop = Transforms.functional.resize(img_crop, size = 704)

    map_crop = Transforms.functional.to_pil_image(Maps)
    map_crop = Transforms.functional.pad(map_crop,512,padding_mode='reflect')
    map_crop = Transforms.CenterCrop(5632)(map_crop)
    map_crop = Transforms.functional.resize(map_crop, size = 704)
    
    img_name = '/content/content/My Drive/ECE4179 - Machine Learning and Neural Networks/ECE4179 Project/Data/Image_crops4/{}.png'.format(image.split('.')[0])
    map_name = '/content/content/My Drive/ECE4179 - Machine Learning and Neural Networks/ECE4179 Project/Data/Map_crops4/{}.png'.format(image.split('.')[0])

#     cv2.imwrite(img_name, img_crop)
    # img_crop.save(img_name)
#     cv2.imwrite(map_name, map_crop) 
    map_crop.save(map_name)
```
