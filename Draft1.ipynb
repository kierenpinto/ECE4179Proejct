```
import torch 
from torch.utils.data import DataLoader

import matplotlib.pyplot as plt
import numpy as np

from unet import unet
from data import dataset
```

```
model = unet()
# model.train()
```

```
trainset = dataset('/content/content/My Drive/ECE4179 Project/Data/Image_crops/', '/content/content/My Drive/ECE4179 Project/Data/Map_crops/', model=model)
```

```
loader = DataLoader(trainset, batch_size=16, num_workers=0)
```

```
batch = next(iter(loader))


f = plt.figure(figsize=(20,20))
for i in range(batch[0].shape[0]):
  ax = f.add_subplot(8, 4,2*i+1)
  ax.imshow(np.moveaxis(batch[0][i].numpy(),0,2))
  ax1 = f.add_subplot(8, 4, 2*i+2)
  ax1.imshow(batch[1][i], cmap='Blues', vmin=0, vmax=5/255)
```
